### [문제11] plt.legend() 함수의 기능은 무엇인가?

```python
font1 = {'family':'serif', 'color':'tan', 'size':20}
font2 = {'family':'serif', 'color':'teal', 'size':14}
font3 = {'family':'serif', 'color':'plum', 'size':14}

y1 = np.array([5, 1, 2, 10, 7, 9])
y2 = np.array([3, 7, 4, 2, 9, 6])
y3 = np.array([4, 6, 8, 6, 4, 5])
plt.plot(y1, label='Python') # 범례에 사용할 라벨 지정
plt.plot(y2, label='ML')
plt.plot(y3, label='DL')

plt.legend() # 범례 설정, 여백에 각 plot의 라벨이 어떤 색깔을 나타내는지를 표시한다.

plt.xlabel("Counts", fontdict=font1)
plt.ylabel("Scores", fontdict=font2)
plt.title("Data Analysis", fontdict=font3)
plt.show()
plt.close()
```

### [실습문제18] Seaborn의 내장 데이터셋인 titanic를 사용하여, 다이아몬드의 나이(age)에 따라 색상지도를 적용한 산점도를 그리시오. x-축은 나이(age), y-축은 운임(fare)으로 설정하고, 나이(age)에 따른 색상지도를 적용하시오. 색상지도는 coolwarm를 사용한다.

답) 승객의 나이(age)에 따라 색상지도를 적용하고, x- 축은 나이(age), y- 축은 운임(fare)로 설정하고, 나이(age)에 따른 색상 지도를 그려 보자.

import matplotlib.pyplot as plt

```python
# 1단계: Seaborn의 내장 데이터셋 'titanic' 불러오기
import matplotlib.pyplot as plt
import seaborn as sns
titanic = sns.load_dataset("titanic")

# 2단계 : 결측치 제거
titanic = titanic.dropna(subset=['age', 'fare'])

# 3단계: scatter plot에 색상지도 적용 ('age'에 따른 색상 적용)

plt.figure(figsize=(8, 6))
print('그래프의 크기 조정을 먼저 해야 새 figure가 생성되지 않는다. \nplt.figure(figsize=(8,6))을 아래로 내리면 그래프가 생성된 이후에 새 figure를 생성하고 그 figure의 크기를 조정한 후, 해당 figure에 컬러맵이 추가되어 경고가 발생한다.')
scatter = plt.scatter(titanic['age'], titanic['fare'],
c=titanic['age'], cmap='coolwarm',
s=100, edgecolor='k')


# 4단계: 색상 범례(색상 막대) 추가
plt.colorbar(scatter, label='Age')

#5단계: 그래프 제목 및 x-축, y-축 레이블 설정
plt.title('Age vs Fare with Colormap')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
plt.close()
```

### [실습문제19] 데이터셋 taxis에서 fare 컬럼을 사용하여 택시 요금(fare)의 분포를 나타내는 히스토그램을 그리시오. 구간의 개수는 12개로 설정하고, 그래프의 색상, 제목, 축 레이블은 여러분이 임의로 추가하시오. 격자선은 y-축만 그린다.

답)

1단계: 데이터셋 taxis 불러오기

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
taxis = sns.load_dataset('taxis')
```

2단계: 히스토그램 그리기

```python
plt.hist(taxis['fare'], bins=12, color='skyblue', edgecolor='navy')

# 그래프 제목 및 축 레이블 설정
plt.title('Taxi Fares', fontsize=16)
plt.xlabel('Fare', fontsize=14)
plt.ylabel('Samples', fontsize=14)

# y축에만 격자선 추가
plt.grid(axis='y', color='olive', linestyle='--')

plt.show()  # 그래프 출력
plt.close()
```

### [실습문제20] [실습문제19 계속] 데이터셋 taxis에서 payment 컬럼을 사용하여 택시요금 결제방법(payment) 비율을 나타내는 파이 차트(Pie chart)를 그리시오. 각 부분에 비율을 소수점 둘째 자리까지 표시하시오. 그래프의 제목은 여러분이 임의로 추가하시오.

답)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
taxis = sns.load_dataset('taxis')
```

1단계: payment 컬럼의 값과 그 개수를 계산한다.

```python
payment_counts = taxis['payment'].value_counts()
print(f'payment_counts is \n \n {payment_counts}')
```

도움말: Pandas의 value_counts() 함수는 가장 기초적이면서 일반적으로 사용되는 함수 중 하나이다.<br>
기본적으로 지정된 열의 각 값(value)에 대한 모든 발생 횟수(빈도수)를 반환한다.<br>
데이터프레임.['컬럼명'].value_counts()<br>

2단계: 파이 차트(Pie chart) 그리기

```python
# 2단계: 파이 차트 그리기
plt.figure(figsize=(8, 6))

colors = ['gold', 'orangered']  # 사용할 색상을 지정

plt.pie(payment_counts, labels=payment_counts.index, autopct='%.2f%%', colors = colors)
plt.title('Taxi Payment methods')
plt.show()
plt.close()
```
