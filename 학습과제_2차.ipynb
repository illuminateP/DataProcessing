### [문제8] tips 데이터셋에서 열 'time'을 기준으로 그룹화한 후,<br> 열 'total_bill'과 'tip'의 평균을 구하려고 한다. 다음에서 올바른 코드를 모두 고르시오.<br>
① df.groupby('time')['total_bill', 'tip'].mean()<br>
② df.groupby('time')[['total_bill', 'tip']].mean()<br>
③ df.groupby('time').agg({'total_bill': 'mean', 'tip': 'mean'})<br>
④ df.groupby('time').mean(['total_bill', 'tip'])<br>

답) 2, 3, 4(예외 처리!)

```python
df = sns.load_dataset('tips')
# print(f"1. Result for df.groupby('time', observed=False)['total_bill', 'tip'] is \n")
# df.groupby('time')['total_bill', 'tip'].mean()
## 인자로 여러 열을 사용할 때는 이중 리스트의 형태로 작성해야 한다.
## alueError: Cannot subset columns with a tuple with more than one element. Use a list instead.

df.groupby('time', observed = False)[['total_bill', 'tip']].mean()
print(f"2. Result for df.groupby('time')[['total_bill', 'tip']].mean() is \n")
display(df.groupby('time', observed = False)[['total_bill', 'tip']].mean())

print(f"3. Result for df.groupby('time').agg({{'total_bill': 'mean', 'tip': 'mean'}}) is \n")
display(df.groupby('time', observed = False).agg({'total_bill': 'mean', 'tip': 'mean'}))
```

```python
print(f".mean()메서드는 일반적으로 특정 열을 인자로 받을 수 없지만, 인자가 주어져도 무시하고 모든 수치형 데이터를 갖는 열의 평균을 출력한다.")
print(f"그러나 이는 권장되는 사용 방식이 아니며, Pandas의 예외 처리 구조 때문에 동작하는 것이다. \n")

print(f"4. Result for df.groupby('time').mean(['total_bill', 'tip'] is \n")
display(df.groupby('time', observed = False).mean(['total_bill', 'tip']))
```

### [문제9] Iris 데이터셋에서 열 'species'을 기준으로 그룹화한 후,<br> 열 'petal_length'의 최솟값과 최댓값을 구하려고 한다. 다음에서 올바른 코드를 모두 고르시오.<br>
① df.groupby('species')['petal_length'].agg(['min', 'max'])<br>
② df.groupby('species').agg({'petal_length': ['min', 'max']})<br>
③ df.groupby('species')[['petal_length']].agg(['min', 'max'])<br>
④ df.groupby('species')['petal_length'].agg({'min', 'max'})<br>

답) 1, 2, 3, 4

```python
df = sns.load_dataset('iris')
df.head()
```

```python
print(f"1. Result for df.groupby('species')['petal_length'].agg(['min', 'max']) is \n")
display(df.groupby('species')['petal_length'].agg(['min', 'max']))

print(f"\n2. Result for df.groupby('species').agg({{'petal_length': ['min', 'max']}}) is \n")
display(df.groupby('species').agg({'petal_length': ['min', 'max']}))
```

```python
print(f"\n3. Result for df.groupby('species')[['petal_length']].agg(['min', 'max']) is \n")
display(df.groupby('species')[['petal_length']].agg(['min', 'max']))

print("★ .agg 함수는 순서가 없는 집합(set)을 인자로 받을 수는 있지만, 이 경우 리스트와 달리 순서가 보장되지는 않는다!")
print(f"4. Result for df.groupby('species')['petal_length'].agg({{'min', 'max'}}) is \n")
display(df.groupby('species')['petal_length'].agg({'min', 'max'}))
```

### [문제11] Iris 데이터셋에서 'sepal_length'와 'sepal_width'를 기준으로 데이터프레임을 병합하려고 한다. 다음에서 올바른 코드를 모두 고르시오. <br>
① pd.merge(df1, df2, on=['sepal_length', 'sepal_width'])<br>
② df1.join(df2, on=['sepal_length', 'sepal_width'])<br>
③ df1.merge(df2, on=['sepal_length', 'sepal_width'])<br>
④ pd.merge(df1, df2, how='outer', on=['sepal_length', 'sepal_width'])<br>

답)1, 3, 4

```python
df = sns.load_dataset('iris')

df1 = df[['sepal_length','sepal_width','petal_width']]
df2 = df[['sepal_length','sepal_width', 'petal_length']]

display(df1)
display(df2)
```

```python
print(f"1. Result for pd.merge(df1, df2, on=['sepal_length', 'sepal_width']) is \n")
display(pd.merge(df1, df2, on=['sepal_length', 'sepal_width']))

# print(f"2. Result for df1.join(df2, on=['sepal_length', 'sepal_width']) is \n")
## ValueError: len(left_on) must equal the number of levels in the index of "right"

print(f"3. Result for df1.merge(df2, on=['sepal_length', 'sepal_width']) is \n")
display(df1.merge(df2, on=['sepal_length', 'sepal_width']))

print(f"4. Result for pd.merge(df1, df2, how='outer', on=['sepal_length', 'sepal_width'])")
display(pd.merge(df1, df2, how='outer', on=['sepal_length', 'sepal_width']))
```
